```python
import tensorflow as pt
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd

# load the dataset
# chuẩn bị dữ liệu (dạng cột hoặc series, dataframe)
dataset = pd.read_csv('data.csv')
print(dataset.shape)
X1 = dataset['kq1'].values
X2 = dataset['kq2'].values
X3 = dataset['kq3'].values
X4 = dataset['kq4'].values
X5 = dataset['kq5'].values
X6 = dataset['kq6'].values

XX = np.column_stack([X1, X2, X3, X4, X5, X6])

X = np.array([XX[0:786,:].T, XX[1:787,:].T, XX[2:788,:].T, XX[3:789,:].T, XX[4:790,:].T]).T
X = X.reshape(786,30)
Y = XX[5:791,:]

print(X[0])

print(X.shape)
print(Y.shape)

# define the keras model
# tạo mô hình AI có 3 lớp
model = Sequential()
model.add(Dense(100, input_shape = (30,), activation='tanh'))      #lớp 1 có 5 NN, hàm kích hoạt dạng tanh()
model.add(Dense(100, activation ='tanh'))                        #lớp 2 có 10 NN, hàm kích hoạt dạng tanh()
model.add(Dense(6, activation = "linear"))                      #lớp ra có 2 NN, hàm kích hoạt tuyến tính y = x

print(model.summary())  #xem tổng quan về mô hình

# compile the keras model
#chỉ tiêu đánh giá MSE (mean squared error), giải thuật huấn luyện "Adam"
model.compile(loss='mse', optimizer='Adam')                    

# fit the keras model on the dataset
#train 10000 lần, 8 dữ liệu được đưa vào trong 1 lần train
model.fit(X, Y, epochs=100, batch_size=16)                    

# evaluate the keras model
#đánh giá mô hình trên một tập dữ liệu khác
X_t = XX[787:792,:].reshape(1,30)
print(X_t)
Y1 = model.predict(X_t) #Dữ liệu Y1 dự báo từ mô hình
print(Y1.round())
```
